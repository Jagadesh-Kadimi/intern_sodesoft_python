```python
import math
class calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def sub(self):
        return self.a-self.b
    def div(self):
        return self.a/self.b
    def mul(self):
      return self.a*self.b
    def modulo(self):
      return math.mod(self.a,self.b)
def main():
  a=int(input("enter first number"))
  b=int(input("enter second number"))
  c=input("enter operator")
  cal=calculator(a,b)
  if c=="+":
     print(cal.add())
  elif c=="-":
     print(cal.sub())
  elif c=="*":
    print(cal.div())
  elif c=="/":
     print(cal.mul())
  elif c=="%":
    print(cal.modulo())
  else:
    print("invalid operator")



if __name__=="__main__":
  main()
```
