```python
class todolist:
   def __init__(self):
        self.todolist = []
   def add_task(self, task):
    self.todolist.append(task)
   def remove_task(self, task):
    self.todolist.remove(task)
   def display_tasks(self):
    print("Your tasks:")
    for task in self.todolist:
      print(task)
   def sort_tasks(self):
    self.todolist.sort()
   def complete_task(self, task):
    self.todolist.remove(task)
    print("Task completed:", task)
    self.display_tasks()
def  main():
    todo = todolist()
    while True:
        print("\n1. Add task\n2. Remove task\n3. Display tasks\n4. Sort tasks\n5. Complete task\n6. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            task = input("Enter task: ")
            todo.add_task(task)
            print("Task added:", task)
        elif choice == "2":
            task = input("Enter task to remove: ")
            todo.remove_task(task)
            print("Task removed:", task)
        elif choice == "3":
            todo.display_tasks
        elif choice == "4":
            todo.sort_tasks()
            print("Tasks sorted.")
        elif choice == "5":
            task = input("Enter task to complete: ")
            todo.complete_task(task)
        elif choice == "6":
            break
        else:
            print("Invalid choice. Please try again.")
if __name__ == "__main__":
    main()
```
